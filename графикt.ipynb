```python
import numpy  as np
import matplotlib.pyplot as plt
data = np.loadtxt('datat.txt')

pi=3.14159265358979323846
e=2.7182818284
x = np.linspace(0, 1)
y = lambda x: np.power(e,-x*pi*pi)
plt.grid()

x1 = data[:, 0]
y1 = data[:, 1]


plt.plot(x1, y1,'r',label='решение для t при x=y=1/2')

plt.plot(x,y(x),'g--',label='точное решение')
plt.legend()
plt.show()
```
